```python
# Ngrams practical implementation
import nltk
from nltk.tokenize import word_tokenize
from nltk import ngrams

def generate_ngrams(text,n):
    token=word_tokenize(text)
    n_grams=list(ngrams(token,n))
    return n_grams
txt='N-grams are a sequence of items from a given sample of text or speech'
unigrams=generate_ngrams(txt,1)
bigrams=generate_ngrams(txt,2)
trigrams=generate_ngrams(txt,3)
print(unigrams)
print(bigrams)
print(trigrams)
```
