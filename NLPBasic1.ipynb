```python
# HANDING CONTRACTIONS USING re LIBRARY
import re
# Function to expand contractions
def expand_contractions(text):
    # Dictionary containing contractions (short forms) and their expansions (full forms)
    contractions_pattern = {
        r"(?i)\bcan't\b": "cannot",      # can't → cannot
        r"(?i)\bwon't\b": "will not",    # won't → will not
        r"(?i)\bI'm\b": "I am",          # I'm → I am
        r"(?i)\bit's\b": "it is",        # it's → it is
        r"(?i)\bthat's\b": "that is",    # that's → that is
    }
    # Loop through each contraction and its expansion
    for contraction, expansion in contractions_pattern.items():
        # Replace all occurrences of the contraction in text with its full form
        text = re.sub(contraction, expansion, text)
    # Return the modified text after all replacements
    return text

# Input text with contractions
text = "I'm happy because I can't go and they don't know what's happening."

expanded_text = expand_contractions(text)
print("Original Text:")
print(text)
print("\nExpanded Text:")
print(expanded_text)
```

```python
# SPECIAL CHARACTER REMOVAL USING re LIBRARY
s="hello how are$ you"
clear_str=re.sub(r"[^a-zA-Z0-9\s]","",s)
print(clear_str)

# Import spaCy
import spacy
print(spacy.__version__)
# Load English language model
nlp = spacy.load('en_core_web_sm')
# Input string
input_str = 'how are% $are you'
# Function to clean text
def clear_text(text):
    # Step 1: Remove special characters (keep only alphabets and spaces)
    cleaned_text = ''.join(char for char in text if char.isalpha() or char.isspace())
    # Step 2: Process text using spaCy
    doc = nlp(cleaned_text)
    # Step 3: Reconstruct sentence from tokens
    return ' '.join(token.text for token in doc)

clear_str = clear_text(input_str)
print("Original Text:")
print(input_str)
print("\nCleaned Text:")
print(clear_str)
```
