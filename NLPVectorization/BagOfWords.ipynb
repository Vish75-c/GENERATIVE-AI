```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

# 1. Load dataset
df = pd.read_csv('spam.csv')

# 2. Convert labels (ham=0, spam=1)
df['spam'] = df['Category'].apply(lambda x: 1 if x == 'spam' else 0)

# 3. Split data
X_train, X_test, Y_train, Y_test = train_test_split(
    df['Message'], df['spam'], test_size=0.2, random_state=42
)

# 4. Convert text → numbers (Bag of Words)
vectorizer = CountVectorizer()

X_train_cv = vectorizer.fit_transform(X_train)   # learn + convert
X_test_cv = vectorizer.transform(X_test)         # only convert
X_train_cv = X_train_cv.toarray()
print(vectorizer.vocabulary_)
# 5. Train model
model = MultinomialNB()
model.fit(X_train_cv, Y_train)

# 6. Predict
y_pred = model.predict(X_test_cv)

# 7. Check accuracy
# print("Accuracy:", accuracy_score(Y_test, y_pred))

# TEST ON A RANDOM DATABASE
message={"I'll call you back in 10 minutes."}
message_cnt=vectorizer.transform(message)
# model.predict(message_cnt)
```
